```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Superstore.csv', sep=';', quotechar='"', encoding='utf-8')
```

```python
from datetime import datetime

def parse_dates(x):
    for fmt in ("%d.%m.%Y", "%m/%d/%Y"):
        try:
            return datetime.strptime(x, fmt)
        except:
            continue
    return pd.NaT

df['Order Date'] = df['Order Date'].apply(parse_dates)
```

```python
df = df.dropna()
```

```python
df.isna().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.drop(
    ['Row ID', 'Ship Date','Ship Mode','Customer ID', 'Customer Name', 'Postal Code'],
    axis=1,
    inplace=True
)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace(r'[^\w\s]', '', regex=True)
```

```python
df.columns
```

```python
num_cols = ['sales', 'profit', 'discount']

df[num_cols] = (
    df[num_cols]
    .astype(str)
    .apply(lambda col: col.str.replace(',', '.', regex=False))
)

df[num_cols] = df[num_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df = df.astype({
     'order_id': 'string',
     'segment': 'string',
     'country': 'string',
      'city': 'string',
      'state': 'string',
      'region': 'string',
      'product_id': 'string',
       'category': 'string',
       'subcategory': 'string',
       'product_name': 'string',
        'sales': 'float64',
        'quantity': 'Int64',
        'discount': 'float64',
       'profit': 'float64'
})
```
